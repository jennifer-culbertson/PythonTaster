We probably want a *visual* sense of how our data is distributed. For example, we can look to see whether reaction time data look as expected, or have lots of outlier (e.g., very long response times):

Ok, now recall that the Stroop Task uses two types of trials, _congruent_ and _incongruent_, to show how automatic processing of the text interferes with color naming. So we want to actually see that effect.  

We're going to do this piece-by-piece so you can get some practice. And, we are going to first use boxplots.

# Appendix. Making interactive plots

The multi-platform plotly library makes it very easy to create interactive plots that can be exported as html, included in web pages, html-based slide presentations, and so on. They are also very useful for exploring your data, as you can hover and immediately see information about each datapoint.

Plotly is not installed on this server, so you'll need to install it for this session:

Load the libraries:

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import pandas as pd
init_notebook_mode(connected=True)
```

```python
# Import the data
expdata = pd.read_csv('ExperimentData.csv')   
```

Let's make a simple scatterplot. There are a number of different ways to create plotly plots in Python. This approach relies on defining a "data" object, which specifies the data to be plotted, and the type of plot to create, and optionally a "layout" object, which specifies things like axes.

```python
data = [ go.Scatter(
        x=expdata['participant'],
        y=expdata['reactionTime'], 
        mode = 'markers',
        marker = dict(      # optionally define marker color and size
            size = 7,
            color = 'rgba(0, 0, 0, .2)'
        )
)]

# optional additional layout options:
layout = go.Layout(
    title="Reaction times, all participants, all trials",
    xaxis=dict(title='participant number'),
    yaxis=dict(title='reaction time'),
    hovermode = 'closest'
)
fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

Looks like participant 4 and 40 had some of the longest outlier reaction times.

Comparing groups would be easy with boxplots:

```python
data = [ 
    go.Box(y = expdata["reactionTime"][expdata['group']=="L1"], name="group 1"
          ),
    go.Box(y = expdata["reactionTime"][expdata['group']=="L2"], name="group 2"
           )
]

layout = go.Layout(title="Two groups", yaxis=dict(title='reaction time'))

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

We could investigate how a single participant performed throughout the experiment (we'll assume the reaction time data is ordered by trials for each participant). Let's also highlight the trials where the text and color are incongruent with bigger markers.

```python
# Subset the data:
p1 = expdata[expdata.participant==2]
# Create some custom labels by pasting together columns:
labels = "text: '" + p1['text'] + "'<br>" + "letter color: " + p1['letterColor']

data = [ go.Scatter(
        x = list(range(1,p1.shape[0]+1)),
        y = p1['reactionTime'], 
        mode = 'lines+markers',
        text = labels,                      # the custom hover labels
        line = dict(color="lightgray"),     # make the lines lighter
        marker = dict(color="black",        # black markers with conditional size:
                      size=p1['congruent'].map({'cong': 7, 'incong': 12})
                     )
    )
       ]

# optional additional layout options:
layout = go.Layout(title="Reaction time data for participant 2",
                yaxis=dict(title='reaction time'),
                xaxis=dict(title='trials')
                  )
fig = go.Figure(data=data, layout=layout)

iplot(fig)
```

So is there a difference between congruent and incongruent trials? In addition to a boxplot, we could investigate a histogram.

```python
g1 = expdata[expdata.group=='L1'] # let's look at group 1

trace1 = go.Histogram(
    x = g1['reactionTime'][g1.congruent=='cong'],
    opacity=0.75, name="congruent"
)
trace2 = go.Histogram(
    x = g1['reactionTime'][g1.congruent=='incong'],
    opacity=0.75, name="incongruent"
)

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',  xaxis=dict(title='reaction time'))
fig = go.Figure(data=data, layout=layout)

iplot(fig, filename='overlaid histogram')
```
